
# Phosphorus Extractors Analysis

A small exploratory notebook based on the phosphorus-extractor paper from the São Francisco Valley.

This notebook is designed as a **portfolio project**, not a full paper reproduction.



## Goals

- Load a compact dataset derived from the paper
- Compare phosphorus extraction values across methods
- Explore how pH, exchangeable Ca, clay, and P-rem relate to the extractors
- Build a short interpretation suitable for a GitHub portfolio


```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/sample_phosphorus_data.csv")
df
```

```python

df.describe().round(2)
```


## Extractor comparison by soil


```python

extractor_cols = ["P_Mehlich1_mg_dm3", "P_Mehlich3_mg_dm3", "P_AER_mg_dm3", "P_Olsen_mg_dm3"]

ax = df.set_index("soil")[extractor_cols].plot(kind="bar", figsize=(10, 6))
ax.set_title("Phosphorus extracted by method across soils")
ax.set_xlabel("Soil")
ax.set_ylabel("P extracted (mg dm$^{-3}$)")
ax.legend(["Mehlich-1", "Mehlich-3", "AER", "Olsen"])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```


## Correlation matrix


```python

soil_cols = ["pH_H2O", "exchangeable_Ca_mmolc_dm3", "clay_g_kg", "P_rem_mg_L"]
corr = df[soil_cols + extractor_cols].corr().round(2)
corr
```


## Mehlich-3 vs soil pH


```python

plt.figure(figsize=(8, 5))
plt.scatter(df["pH_H2O"], df["P_Mehlich3_mg_dm3"])
for _, row in df.iterrows():
    plt.annotate(row["soil"], (row["pH_H2O"], row["P_Mehlich3_mg_dm3"]), xytext=(5, 5), textcoords="offset points")
plt.title("Mehlich-3 extraction vs soil pH")
plt.xlabel("Soil pH")
plt.ylabel("Mehlich-3 (mg dm$^{-3}$)")
plt.tight_layout()
plt.show()
```


## Mehlich-1 vs exchangeable calcium


```python

plt.figure(figsize=(8, 5))
plt.scatter(df["exchangeable_Ca_mmolc_dm3"], df["P_Mehlich1_mg_dm3"])
for _, row in df.iterrows():
    plt.annotate(row["soil"], (row["exchangeable_Ca_mmolc_dm3"], row["P_Mehlich1_mg_dm3"]), xytext=(5, 5), textcoords="offset points")
plt.title("Mehlich-1 extraction vs exchangeable Ca")
plt.xlabel("Exchangeable Ca (mmolc dm$^{-3}$)")
plt.ylabel("Mehlich-1 (mg dm$^{-3}$)")
plt.tight_layout()
plt.show()
```


## Mehlich-1 / Mehlich-3 ratio


```python

df["M1_to_M3_ratio"] = df["P_Mehlich1_mg_dm3"] / df["P_Mehlich3_mg_dm3"]
plt.figure(figsize=(8, 5))
plt.bar(df["soil"], df["M1_to_M3_ratio"])
plt.title("Mehlich-1 / Mehlich-3 ratio by soil")
plt.xlabel("Soil")
plt.ylabel("Ratio")
plt.tight_layout()
plt.show()

df[["soil", "M1_to_M3_ratio"]].round(2)
```


## Interpretation

- **Soil V** is an outlier because it combines high pH, high exchangeable Ca, and high clay.
- **Mehlich-1** shows more erratic behavior across soils.
- **Mehlich-3** appears more stable in this compact dataset.
- This matches the broader conclusion of the paper: in alkaline environments, exchangeable calcium and pH matter strongly, and Mehlich-3 performs well.
